```python
import pandas as pd
```

```python
df=pd.read_csv('/content/diabetes_young_adults_india.csv')
```

```python
df
```

```python
print("Summary Statistics:")
print(df.describe())
```

```python
correlation_matrix = df.select_dtypes(include=['number']).corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
bmi_distribution = df['BMI'].describe()
print("\nDistribution of BMI:")
print(bmi_distribution)
```

```python
fasting_blood_sugar_stats = df.groupby('Diabetes_Type')['Fasting_Blood_Sugar'].describe()
print("\nFasting Blood Sugar by Diabetes Type:")
print(fasting_blood_sugar_stats)
```

```python
q1 = df['HbA1c'].quantile(0.25)
q3 = df['HbA1c'].quantile(0.75)
iqr = q3 - q1
outliers = df[(df['HbA1c'] < (q1 - 1.5 * iqr)) | (df['HbA1c'] > (q3 + 1.5 * iqr))]
print("\nOutliers in HbA1c:")
print(outliers[['ID', 'HbA1c']])
```

```python
physical_activity_counts = df['Physical_Activity_Level'].value_counts()
print("\nPhysical Activity Level Distribution:")
print(physical_activity_counts)
```

```python
selected_features = ['BMI', 'Fasting_Blood_Sugar', 'HbA1c', 'Cholesterol_Level', 'Genetic_Risk_Score']
feature_correlations = df[selected_features].corr()
print("\nPairwise Correlations for Selected Features:")
print(feature_correlations)
```
